```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
train_set = pd.read_csv('../train.csv')
test_set = pd.read_csv('../test.csv')
```

```python
unique_id = train_set.place_id.unique()
random_index = np.random.choice(len(unique_id), 50)
```

```python
checkins = train_set[train_set.place_id == unique_id[random_index[0]]]
plt.figure(figsize=(15, 2))
plt.xlim([0,10080])
plt.plot(checkins.time % (60 * 24 * 7),np.zeros_like(checkins.time), 'ro')
```
