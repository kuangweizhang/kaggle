```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
train_set = pd.read_csv('../train.csv')
test_set = pd.read_csv('../test.csv')
```

```python
import math
```

```python
unique_id = train_set.place_id.unique()
random_index = np.random.choice(len(unique_id), 50)
```

```python
for index in range(0,49):
    checkins = train_set[train_set.place_id == unique_id[random_index[index]]]
    plt.figure(figsize=(15, 2))
    plt.xlim([0,10080])
    plt.xticks(np.arange(0,10080,1440))
    plt.plot((checkins.time + 120) % (60 * 24 * 7),np.zeros_like(checkins.time), 'ro')
    plt.show()
    
    days = np.floor(((checkins.time + 120)/(60 * 4)))%(42)
    counts = days.value_counts()
    counts.sort_index(inplace=True)
    plt.ylim([0,300])
    plt.xlim([0,42])
    plt.xticks(np.arange(0,42, 6))
    plt.bar(counts.index, counts.values, 0.35)
    plt.show()
    
    print(index)
```
