```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
train_set = pd.read_csv('../../train.csv')
```

```python
train_set.accuracy.describe()
```

```python
unique_id = train_set.place_id.unique()
random_index = np.random.choice(len(unique_id), 50)
```

```python
for index in range(0,49):    
    checkins = train_set[train_set.place_id == unique_id[random_index[index]]]
    
    print(len(checkins.index))
    
    plt.figure(figsize=(5, 5))
    plt.ylim([0,10])
    plt.xlim([0,10])
    plt.plot(checkins.x, checkins.y,'bo')
    veryaccurate = checkins[checkins.accuracy > 27]
    plt.plot(veryaccurate.x, veryaccurate.y,'ro')
    plt.show()
    
    plt.figure(figsize=(5, 5))
    plt.ylim([0,10])
    plt.xlim([0,10])
    plt.plot(checkins.x, checkins.y,'bo')
    veryaccurate = checkins[checkins.accuracy > 62]
    plt.plot(veryaccurate.x, veryaccurate.y,'ro')
    plt.show()
    
    plt.figure(figsize=(5, 5))
    plt.ylim([0,10])
    plt.xlim([0,10])
    plt.plot(checkins.x, checkins.y,'bo')
    veryaccurate = checkins[checkins.accuracy > 75]
    plt.plot(veryaccurate.x, veryaccurate.y,'ro')
    plt.show()
    
    print(index, '=====================')
```
