```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
bids = pd.read_csv('./../data/bids.csv')
train = pd.read_csv('./../data/train.csv')
test = pd.read_csv('./../data/test.csv')
sample_submission = pd.read_csv('./../sampleSubmission.csv')
```

```python
# Bots Ips are not used by non bots ips
bids_merged = bids.merge(train,
          how='left',
          on='bidder_id')
```

```python
bids_merged.head()
```

```python
bids_train = bids_merged[~np.isnan(bids_merged.outcome)]
```

```python
bids_train
```

```python
bot_bids = bids_train[bids_train.outcome==1]
```

```python
bot_bids.sort_values(by=['bidder_id','auction'])
```

```python
normal_bids = bids_train[bids_train.outcome==0]
```

```python
# Check if payment account,address to bidder id is one to one mapping
print(len(train.index))
print(train.bidder_id.nunique())
print(train.payment_account.nunique())
print(train.address.nunique())
```

```python
# Number of total bids compare is a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['bid_id'].agg(['count']).describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['bid_id'].agg(['count']).describe())
```

```python
# Number of merchandis is not a feature
print(normal_bids.groupby(['bidder_id'])['merchandise'].nunique().describe())
print(bot_bids.groupby(['bidder_id'])['merchandise'].nunique().describe())
```

```python
# number of device per bidder is a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['device'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['device'].nunique().describe())
```

```python
# Numbers of auctions is a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['auction'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['auction'].nunique().describe())
```

```python
# Number of countries is a feature
print(normal_bids.groupby(['bidder_id'])['country'].nunique().describe())
print(bot_bids.groupby(['bidder_id'])['country'].nunique().describe())
```

```python
# Number of payment accounts is not a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['payment_account'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['payment_account'].nunique().describe())
```

```python
# Number of addresses compare is not a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['address'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['address'].nunique().describe())
```

```python
# Compare if number of ip is a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id'])['ip'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id'])['ip'].nunique().describe())
```

```python
# Number of address is not a feature
print('normal bids')
print(normal_bids.groupby(['bidder_id','auction'])['address'].nunique().describe())
print()
print('bot bids')
print(bot_bids.groupby(['bidder_id','auction'])['address'].nunique().describe())
```
