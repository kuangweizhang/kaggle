### Conclusion of data exploring
Everything looks fine except there is a special coordinate -120.5 and 90. This looks like a default coordinate when GPS is not working. try to recover the coordinate

#### Able to recover some coordinates from test/train data

```python
# How many addressing has the missing GPS
missing_address_rows = total_set[(total_set.Y == 90)&(total_set.X == -120.5)]
len(missing_address_rows.index)
```

```python
address_coordinate = total_set[
    (total_set.Address.isin(missing_address_rows.Address)) & 
    (total_set.Y != 90) & 
    (total_set.X != -120.5)]
len(address_coordinate.index)
```

```python
for index, row in missing_address_rows.iterrows():
    for index2, row2 in address_coordinate.iterrows():
        if(row.Address == row2.Address and row.X == -120.5 and row.Y == 90):
            total_set.loc[index, 'X'] = row2.X
            total_set.loc[index, 'Y'] = row2.Y
```

```python
len(total_set[(total_set.Y == 90)&(total_set.X == -120.5)].index)
```

#### Not very significant, but recovered some of the GPS

### SVM

```python
train_set = total_set[np.isnan(total_set.Id)]
```
