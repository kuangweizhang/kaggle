```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
train = pd.read_json('./input/train.json')
test = pd.read_json("./input/test.json")

from sklearn.model_selection import train_test_split
X_train, X_validate, y_train, y_validate = train_test_split(\
    train[['bathrooms', 'bedrooms','price','latitude','longitude']], \
    train[['interest_level']].values, \
    test_size=0.33, random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC(probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
predict_validate = clf.predict_proba(X_validate)
from sklearn.metrics import log_loss
print(log_loss(y_validate, predict_validate))
```

```python
result = clf.predict_proba(test[['bathrooms', 'bedrooms','price','latitude','longitude']])
result_frame = pd.DataFrame(result)
result_frame.columns = clf.classes_
result_frame['listing_id'] = test.listing_id.reset_index().listing_id
```

```python
result_frame[['listing_id','high','medium','low']].to_csv('result.csv', \
                                                          index = False)
```
